```python
x = df.drop(columns=['Salary'])
y = df["Salary"]
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_text,y_train,y_text = train_test_split(x,y,test_size = 0.2)
```

```python
len(x_train)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model
```

```python
#Training data
model.fit(x_train,y_train) #fit is afunction that takes input of training data and will gives output of that
```

```python
#test

y_pred=model.predict(x_text) #predict is a function that test given input
```

```python
pd.DataFrame({'y_test':y_text,'y_pred':y_pred,"MAE":y_text-y_pred})
```

```python
#import metrics
import numpy as np
from sklearn.metrics import mean_squared_error,r2_score,mean_absolute_error
print("Mean absolute error:",mean_absolute_error(y_text,y_pred))
print("Mean square root",mean_squared_error(y_text,y_pred))
print("Root Mean square root",np.sqrt(mean_squared_error(y_text,y_pred)))
print("R2-square",r2_score(y_text,y_pred))
```
